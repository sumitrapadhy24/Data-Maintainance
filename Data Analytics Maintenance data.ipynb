```python
con=["lifetime","pressureInd","moistureInd","temperatureInd"]
cat=["broken","team","provider"]
```

```python
#univariate
#bivariate
#multivariate
```

```python
for i in cat:
    sns.countplot(x=i,data=df)
    plt.show()
    
```

```python
for i in con:
    sns.distplot(df[i])
    plt.show()
```

```python
for i in con:
    sns.scatterplot(x=df.index, y=df[i])
    plt.show()
```

```python
#bivariate analysis
for i in con:
    sns.swarmplot(x="broken",y=i,data=df)
    plt.show()
```

```python
for i in con:
    sns.scatterplot(df[i],df["lifetime"])
    plt.show()
```
